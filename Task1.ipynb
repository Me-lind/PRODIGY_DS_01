```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file, skipping the first 4 lines and specifying the correct delimiter
df = pd.read_csv('population.csv', sep=',', skiprows=4, encoding='utf-8')

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# Display basic information about the DataFrame
print(df.info())

# Display summary statistics
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())

# drop missing values
df = df.dropna()
```

```python
# Plot population growth for a specific country
country_name = 'Angola'
country_data = df[df['Country Name'] == country_name].iloc[0, 4:]  # Select columns for the years
years = country_data.index
population = country_data.values

plt.figure(figsize=(10, 5))
plt.plot(years, population, marker='o')
plt.title(f'Population Growth in {country_name}')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

```python
# Calculate total population for each year
total_population_per_year = df.iloc[:, 4:].sum()
print(total_population_per_year)

# Plot the total population over the years
plt.figure(figsize=(10, 5))
plt.plot(total_population_per_year.index, total_population_per_year.values, marker='o')
plt.title('Total World Population Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.grid(True)
plt.show()
```
